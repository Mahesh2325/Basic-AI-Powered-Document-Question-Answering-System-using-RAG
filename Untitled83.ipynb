```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("/content/AI_QA_Test (1).pdf")
documents = loader.load()

print("Total pages:", len(documents))
print(documents[0].page_content[:500])  # preview first page
```

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import CharacterTextSplitter



text_splitter = CharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)

docs = text_splitter.split_documents(documents)

print("Total chunks:", len(docs))
print(docs[0].page_content)
```

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_community.embeddings import HuggingFaceEmbeddings



text_splitter = CharacterTextSplitter(chunk_size=500, chunk_overlap=50)
docs = text_splitter.split_documents(documents)

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2",
    model_kwargs={"device": "cpu"}
)

# Test embedding
vector = embeddings.embed_query("What is AI?")

print("Vector length:", len(vector))
```

```python
from langchain_community.vectorstores import FAISS
vectorstore = FAISS.from_documents(docs, embeddings)
```

```python
query = "What is AI?"

results = vectorstore.similarity_search(query, k=2)

for i, doc in enumerate(results):
    print(f"\nResult {i+1}:")
    print(doc.page_content[:300])
```

```python
from transformers import pipeline

# Load model
from transformers import pipeline

pipe = pipeline(
    "text-generation",
    model="gpt2"
)
# Chat loop
while True:
    query = input("\nAsk: ")

    if query.lower() == "exit":
        break

    # Retrieve
    results = vectorstore.similarity_search(query, k=1)

    context = ""
    for doc in results:
        context += doc.page_content + "\n"

    # Better prompt (IMPORTANT)
    prompt = f"""
Answer the question clearly.

Context:
{context}

Question: {query}

Answer:
"""

    response = pipe(
        prompt,
        max_new_tokens=80,
        truncation=True,
        do_sample=True,
        temperature=0.7
    )

    full_text = response[0]['generated_text']

    # ✅ Extract only answer part
    if "Answer:" in full_text:
        answer = full_text.split("Answer:")[-1]
    else:
        answer = full_text

    print("\nAnswer:", answer.strip())
```
